```python
#4
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
```

```python
#5
df['depth'].iloc[42]
```

```python
#6
df[df['color']=='E']
```

```python
#7
df['purchase']=np.nan
z=0
for i,j in zip(df['carat'].to_list(),df['cut'].to_list()):
    if i>.4 and j=='Premium':
        df['purchase'].iloc[z]='Yes'
    else:
        df['purchase'].iloc[z]='No'
    z=z+1
      
```

```python
#8
df['carat'].median()
```

```python
#9
df['price'].mean()
```

```python
#10
(df['x']*df['y']*df['z']).mean()
```

```python
#11
df=df.drop(df[df['cut']=='Fair'].index.to_list(),axis='index')
df
```

```python
df.sort_values(by="carat",inplace=True)
df
```
