```python
import pandas as pd
```

```python
df=pd.read_csv("fred_data.csv")
df.columns=['date','gdp','investment','consumption']
```

```python
df.set_index("date",inplace=True)
df=df.loc['1/1/2000':'10/1/2009']
```

```python
for i in df.columns.to_list():
    temp=df[i].pct_change().to_list()
    df[(i+" growth")]=temp
```

```python
df=df[['gdp', 'gdp growth', 'investment','investment growth', 'consumption', 'consumption growth']]
df.fillna(int(0),inplace=True)
df.head()
```

```python
#highest gdp growth val
df.index.to_list()[(df['gdp growth'].to_list()).index(df['gdp growth'].min())]
```

```python
#Write a function to label a quarter "recession" if the growth rate of a variable is negative, and "normal" otherwise.

def label(quarter):
    quarter=str(quarter)
    rates=df[['gdp growth', 'investment growth', 'consumption growth',]].loc[quarter].to_list()
    
    if rates[0]<0:
        return "recession"

    elif rates[1]<0:
        return "recession"
    
    elif rates[2]<0:
        return "recession"
    
    else:
        return "normal"
label('10/1/2008')
```

```python
df['quarter']=[label(i) for i in df.index.to_list()]
df.head()
```

```python
#alternate method using applymaps
def label2(rate):
    if rate>0:
        return "normal"
    else:
        return "recession"
df[['gdp growth', 'investment growth', 'consumption growth',]].iloc[1:].applymap(label2).head()
```
