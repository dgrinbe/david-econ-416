```python
import pandas as pd
df=pd.read_csv("Binder_Access_Spring.csv")
```

```python
df.head(2)
```

```python
df.columns=['date','count']
df.tail(3)
```

```python
len(df)
```

```python
df['date']=pd.to_datetime(df['date'])
df.dtypes
```

```python
df.set_index('date',inplace=True)
df.head()
```

```python
print(df['count'].sum())
print(df['count'].mean())
print(df['count'].median())
```

```python
df['count']
```

```python
df[df['count']>0]
```

```python
median=df['count'].median()
mean=df['count'].mean()


import matplotlib.pyplot as plt

fig, ax=plt.subplots(1,2,figsize=(30,14))

ax[0].plot(df.index,df['count'].to_list(),color='green',label="Binder Access",linewidth=4)
ax[0].set_title("Number of Times the Binder was accessed Spring")
ax[0].set_xlabel("Date accessed")
ax[0].set_ylabel("Times Accessed")
ax[0].set_yticks(list(range(0,130,10)))
ax[0].set_xticks(ticks=['2022-02-01','2022-02-15','2022-03-01','2022-03-15','2022-04-01','2022-04-15','2022-05-01','2022-05-15','2022-05-25'],
                 labels=['feb 1','feb 15','mar 1','mar 15','apr 1','apr 15','may 1','may 15','may 25',])

for i in list(range(0,130,10)):
    ax[0].axhline(i,color='black',alpha=.3,linestyle='--')

ax[0].axhline(median,color='orange',linestyle='--',label="median")
ax[0].axhline(mean,color='red',linestyle='--',label="mean")

    
ax[0].legend()

ax[1].hist(df['count'].to_list(),bins=(list(range(0,130,5))),align='mid',color='green',edgecolor='black')

ax[1].set_title("Histogram of daily Binder access Spring")
ax[1].set_xlabel("Value")
ax[1].set_ylabel("# of times value occured")
ax[1].set_yticks(list(range(0,75,5)))
#ax[1].set_xticks(
for i in list(range(0,70,5)):
    ax[1].axhline(i,color='black',alpha=.3,linestyle='--')

ax[1].axvline(median,color='orange',linestyle='--',label="median")
ax[1].axvline(mean,color='red',linestyle='--',label="mean")
ax[1].legend()

plt.show()
```
